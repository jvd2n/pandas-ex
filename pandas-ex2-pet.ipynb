```python
# 4. Create DataFrame
data = {'animal': ['cat', 'cat', 'snake', 'dog', 'dog', 'cat', 'snake', 'cat', 'dog', 'dog'],
        'age': [2.5, 3, 0.5, np.nan, 5, 2, 4.5, np.nan, 7, 3],
        'visits': [1, 3, 2, 3, 2, 3, 1, 1, 2, 1],
        'priority': ['yes', 'yes', 'no', 'yes', 'no', 'no', 'no', 'yes', 'no', 'no']}
labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j']
```

```python
a = pd.DataFrame(data, index=labels)
```

```python
# 5. df Output
a
```

```python
# 6. 객체 상위 3열까지 출력하시오
a.head(3)
```

```python
# 7. animal과 age 컬럼만 출력
a.loc[:, ['animal','age']]
```

```python
# 8. 3,4,8열에 해당하는 animal과 age 출력
a.loc[[a.index[3], a.index[4], a.index[8]], ['animal', 'age']]
```

```python
# 9. visits 컬럼에서 3 초과하는 값 출력
a[a['visits'] >= 3]
```

```python
# 10. age에서 NaN 값 출력
a[a['age'].isnull()]
```

```python
# 11.age가 3살 미만 고양이값 출력
a[(a['age'] <= 3) & (a['animal'] == 'cat')]
```

```python
# 12, age가 2살 이상 4실 미만인 값 출력
a[(a['age'] >= 2) & (a['age'] < 4)]
```

```python
# 13. f행의 나이를 1.5살로 변경
# a.iloc[0][1] = 1.5
# a
a.loc['f', 'age'] = 1.5
a
```

```python
# 14. visits의 합 출력
a['visits'].sum()
```

```python
# 15. 동물별로 나이의 평균 출력
a.loc[:, ['animal', 'age']].groupby(['animal'], as_index=False).mean()
```

```python
# 16. k행을 추가하여 dog, 5.5, 2, no 추가
a
a.append({'animal':'dog', 'age':5.5, 'visits':2, 'priority':'no'}, ignore_index=True)
```

```python
a.loc['k'] = ['dog', 5.5, 2, 'no']
a
```

```python
# 16. 추가한 행 삭제
a.drop('k', axis=0)
```

```python
# 17. 객체에 있는 동물의 종류의 수 출력
a['animal'].value_counts()
```

```python
# 18. age는 내림차순, vistis는 오름차순 정렬
a.sort_values(by=['age', 'visits'], ascending=[False,True])
```

```python
# 19. priority의 yes를 True, no를 False로 매핑 후 출력
a['priority'] = a['priority'].map({'yes': True, 'no': False})
a
```

```python
# 20. snake를 python으로 값 변경
a['animal'] = a['animal'].replace('snake', 'python')
a
```

```python
# 21. 각각의 동물의 유형과 방문 횟수에 대해 평균 나이를 찾으시오.
# 각 행은 animal 이고, 각 열은 visit 이며 값은 평균 나이
a.pivot_table(index='animal', columns='visits', values='age', aggfunc='mean')
```
